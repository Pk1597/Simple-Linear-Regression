Importing Libraries and Reading Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
import statsmodels.api as sm
dataset = pd.read_csv('Salary_Data.csv')
dataset.head()
```

Plotting data

```python
plt.figure(figsize=(16, 8))
plt.scatter(
    dataset['YearsExperience'],
    dataset['Salary'])
plt.xlabel("Years of Experience")
plt.ylabel("Salaries")
plt.show()
```

Linear Fitting Data

```python
X = dataset['YearsExperience'].values.reshape(-1,1)
y = dataset['Salary'].values.reshape(-1,1)
reg = linear_model.LinearRegression()
reg.fit(X, y)
print("The linear model is: Y = {:.5} + {:.5}X".format(reg.intercept_[0], reg.coef_[0][0]))
```

Plotting the Data after fitting

```python
predictions = reg.predict(X)
plt.figure(figsize=(16, 8))
plt.scatter(
    dataset['YearsExperience'],
    dataset['Salary'],
    c='black'
)
plt.plot(
    dataset['YearsExperience'],
    predictions,
    c='blue',
    linewidth=2
)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

Printing the Statistical Summary

```python
X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```

Testing with the year of experience

```python
Years_experience = input("Enter the Years of experience : ")
Years_experience = int(Years_experience)
Salary_employee = reg.intercept_[0] + (reg.coef_[0][0]*Years_experience)
print(Salary_employee)
```
